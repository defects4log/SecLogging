```python
import pandas as pd

label_data_path1 = "./log_security/results/Datas/combined_result_test.csv"
label_data_path2 = "./log_security/results/Datas/combined_result.csv"
def read_df(path):
    encodings_to_try = ['utf-8-sig', 'utf-8', 'gbk', 'latin-1']
    for encoding in encodings_to_try:
        try:
            df = pd.read_csv(path, encoding=encoding)
            print(f"Successfully read {path} with encoding: {encoding}")
            return df
        except UnicodeDecodeError:
            print(f"Failed to read {path} with encoding: {encoding}")
    raise ValueError(f"Could not read the file {path} with tried encodings.")

def extract_data(model_name, method):
    merged_res = f"./log_security/results/different_model/{model_name}/merged/{method}__merged_similarity_analysis_results.csv"

    df1 = read_df(label_data_path1)
    df2 = read_df(label_data_path2)
   
    
    title_to_data = {}

    for _, row in pd.concat([df1, df2]).iterrows():
        title_to_data[row['Issue Title']] = row
    
    print(f"Total unique titles in label data: {len(title_to_data)}")

    df_merged = pd.read_csv(merged_res)

    res_headers = ['Title', 'BeforeContent', 'AfterContent', 'fixed_code', 'author1_judge', 'author2_judge', 'final_label']
    res_data = []
    for _, row in df_merged.iterrows():
        title = row['issue_title']
      
        if title in title_to_data:
            label_row = title_to_data[title]
            res_data.append([
                title,
                label_row['BeforeContent'],
                label_row['AfterContent'],
                row['fixed_code_content'],
                "",
                "",
                ""
            ])
        else:
            print(f"Title not found in label data: {title}")
    
    print(f"Total extracted rows: {len(res_data)}")
    res_df = pd.DataFrame(res_data, columns=res_headers)
    output_path = f"./log_security/results/different_model/{model_name}/{method}_for_labeling.csv"
    res_df.to_csv(output_path, index=False)
    print(f"Extracted data saved to {output_path}")

extract_data("deepseek-r1_32b", "with_description")
    
```
